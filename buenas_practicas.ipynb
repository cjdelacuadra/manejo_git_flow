```python
def main(top, raw_tweets):
    if top == 'tweets':
        fn = top_retweeted
    elif top == 'usuarios':
        fn = top_usuarios
    elif top == 'dias':
        fn = top_dias
    elif top == 'hashtags':
        fn = top_hashtags
    return fn(raw_tweets)
```

```python
def top_retweeted(tweets):
    top = tweets.nlargest(10, ['retweetCount'])
    return top
```

```python
def top_usuarios(tweets):
    users = json_normalize(tweets['user'])
    users.drop(['description', 'linkTcourl', 'friendsCount', 'followersCount', 
            'verified', 'descriptionUrls', 'profileImageUrl', 'protected', 
            'location', 'profileBannerUrl', 'url', 'rawDescription', 'favouritesCount', 'linkUrl'], axis=1, inplace=True)
    top = users.nlargest(10, ['statusesCount'])
    return top
```
